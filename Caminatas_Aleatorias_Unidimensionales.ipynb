## **Caminatas Aleatorias Unidimensionales**

**Objetivo:** Simular y graficar la trayectoria de una caminata aleatoria unidimensional.

**Definición:** Sea $\{\xi_n\}_{n\geq1}$ una sucesión de variables aleatorias independientes e idénticamente distribuidas (v.a.i.i.d) tal que cada variable $\xi_n$ puede tomar un valor $r$ con prababilidad $p$ o un valor $-t$ con probabilidad $q=1-p$, donde $p\in (0,1)$ y $t,r>0$. Se define entonces una caminata aleatoria unidimensional como el proceso estocástico $\{X_n\}_{n=0}^{\infty}$ dado por $X_n = \sum_{i=1}^n \xi_i$ para cada $n \in \mathbb{N}$ y $X_0 = y$ para algún número real fijo $y$ que representa el punto de partida.

Este proceso representa el número de 'pasos' hacía la derecha de longitud $r$ y de 'pasos' a la izquierda de longitud $t$ que se han dado hasta el tiempo $n$ partiendo del punto $y$.

**Nota:** Sí el proceso $\{X_n\}_{n=0}^\infty$ definido previamente es tal que $r=1=t$ y $p=\frac{1}{2}$, entonces se le llama *caminata aleatoria simple (simétrica) unidimensional*.

Ahora, vamos a simular una caminata aleatoria en su sentido más general (es decir, conforme a le definición inicial dada)

## **Simulación de trayectorias para caminatas aleatorias**

```python
# Paqueterias a utilizar:
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def simular_caminata(n_pasos, n_sims = 1, pos_inicial = 0, l_der = 1, l_izq = 1, p = 0.5):
  '''
  Función para simular una caminata aleatoria unidimensional iniciando en la posición inicial 'pos'.
  Inputs
  Param n_pasos: número de pasos a realizar en la simulación
  Param n_sims: número de simulaciones o escenarios de la caminata
  Param pos_inicial : posición inicial de la caminata
  Param l_der : longitud de los pasos a la derecha (número real positivo)
  Param l_izq : longitud de los pasos a la izquierda (número real positivo)
  Param p : probabilidad de ir a la derecha
  Outputs
  Arreglo numpy con la posición de la caminata aleatoria en cada paso

  Nota 1: Por default, al sólo ingresar el número de pasos, se simulará una caminata aleatoria simple (simétrica) una sóla vez que parte del origen (0)
  Nota 2: En cada simulación se imprime en la consola la posición final de la caminata aleatoria
  '''
  posiciones_sims = []
  for _ in range(n_sims):
    inicial = pos_inicial
    posiciones = np.zeros(n_pasos+1) # Vector que contendrá al posiciones de la caminata en cada tiempo
    posiciones[0] = inicial # Al tiempo 0 estamos en el punto incial 'pos_inicial'.
    for i in range(n_pasos):
      inicial += random.choices([l_der,-l_izq], weights=[p,1-p], k=1)[0]
      posiciones[i+1] = inicial
    posiciones_sims.append(posiciones)
    print(f'La posición de final de la simulación {_+1} es: {round(posiciones[-1],2)}')
  return posiciones_sims
```

```python
def graficar_caminata(posiciones):
  '''
  Función para graficar las simulaciones de una caminata aleatoria unidimensional en una misma gráfica.
  Input
  Param posiciones: arreglo numpy con las posiciones de cada simulación de la caminata aleatoria en cada tiempo
  '''
  x=np.arange(len(posiciones[0]))
  plt.figure(figsize=(10,7))
  colores = plt.cm.tab20(np.linspace(0, 1, len(posiciones)))
  for i, color in zip(posiciones, colores):
    plt.plot(x, i, color=color, alpha =0.7)
  plt.xlabel('Tiempo')
  plt.ylabel('Posición')
  plt.title(f'{len(posiciones)} Simulaciones de una Caminata Aleatoria Unidimensional de {len(posiciones[0])-1} pasos')
  plt.grid(True)
  plt.show()
```

### **Ejemplos Ilustrativos**

```python
# Ejemplo: Gráfica de Caminata aleatoria simple simétrica de 100 pasos (20 escenarios):
graficar_caminata(simular_caminata(n_pasos=100, n_sims=20))
```

```python
# Ejemplo: Gráfica de Caminata aleatoria de 20 pasos que parte del origen con longitud de paso a la derecha de 2, longitud de paso
# a la izquierda de 0.5 y probabilidad de ir a la derecha de 0.4 (30 escenarios):
graficar_caminata(simular_caminata(n_pasos=20, n_sims=30, pos_inicial=0, l_der=2, l_izq=0.5, p=0.4))
```

```python
# Ejemplo: Gráfica de Caminata aleatoria de 50 pasos que parte del punto y=5 con longitud de paso a la derecha de 0.75, longitud
# de paso a la izquierda de 2 y probabilidad de ir a la derecha de 0.7 (15 escenarios):
graficar_caminata(simular_caminata(n_pasos=50, n_sims=15, pos_inicial=5, l_der=0.75, l_izq=2, p=0.7))
```

## **Caso Práctico:**  Precio de un activo financiero según una dinámica binomial aditiva

A continuación vamos a considerar un activo cuya dinámica de precios se comporta como un árbol binomial aditivo, de modo que, cada posible trayectoria del precio es una caminata aleatoria unidimensional. El precio de este activo parte de un precio inicial $S_0$ al día de hoy y vamos a suponer que existen dos constantes $u > 0$ y $d < 0$ tales que el precio en un cierto punto del tiempo (medido de forma discreta) $S_n$ sube sumando $u$ (es decir, $S_u=S_n+u$) y baja sumando $d$ (es decir, $S_d = S_n+d$). Supongamos además que se tiene interés en simular la trayectoria que tomará este activo dentro de 7, 10 y 30 días.

Para fines ilustrativos vamos a considerar el petróleo crudo (Commodity) y vamos a suponer que su precio sigue una dinámica binomial aditiva como la descrita anteriormente. Más aún, vamos a considerar el *Brent Crude Oil Last Day Financ* con clave o ticker *BZ=F* en *Yahoo Finance* que es la referencia internacional para el precio del petróleo crudo.

Lo primero que haremos será obtener la serie histórica de precios usando la libreria 'yfinance' de Python; seguido de esto, buscaremos estimar las constantes $u$ y $d$ y finalmente, haremos la simulación de escenarios para la trayectoria del precio del activo dentro de 7, 10 y 30 días (20 escenarios en total para cada caso).


Vamos a comenzar importando las librerias que utilizaremos en esta parte, así como descargando la información de Yahoo Finance del activo con ticker *'BZ=F'* tomando como fecha inicial el 01 de enero de 2025. Asimismo, vamos a realizar la gráfica de la serie histórica del precio.

```python
# Paqueterias a usar:
import yfinance as yf
import datetime as dt
```

```python
# Descarga de la serie histórica de precios
activo = 'BZ=F'
inicio = dt.datetime(2025,1,1) # Fecha inicial: 01 de enero de 2025
hoy = dt.datetime.today() # Fecha final: ültima observación al momento de ejecutar el código
datos = yf.download(activo, inicio, hoy)['Close'] # Descarga de los datos (serie histórica)
datos.tail()
```

```python
# Gráfica del movimiento histórico del precio desde el 01 de enero de 2025
plt.figure(figsize=(10,7))
plt.plot(datos)
plt.title(f'Serie histórica de precios: {activo}')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.grid(True)
plt.show()
```

```python
# Precio S0 al dia de hoy
S0 = datos.iloc[-1][activo]
print(f'Precio S0: {S0}')
```

Como siguiente paso, vamos a calcular las diferencias en el precio entre cada una de las fechas y la fecha inmediata anterior con la finalidad de estimar la probabilidad de que el precio suba, así como realizar una estimación de las constantes $u$ y $d$.

```python
# Diferencias de precios entre cada fecha y la inmediata anterior
datos = datos.diff()
datos.head()
```

Primeramente, vamos a estimar la probabilidad de que el precio suba como la proporción de todas las diferencias positivas (subidas) entre el total de diferencias.


```python
p=datos[datos>0].count()[activo]/datos.count()[activo]
print(f'Probabilidad de subida en el precio: {p}')
```

Luego, estimaremos las constantes $u$ y $d$ apoyándonos en las diferencias calculadas anteriormente tomando $u =\mu + \sigma$ y $d=\mu - \sigma$, donde $\mu$ y $\sigma$ son la media y desviación estándar muestrales de la serie de diferencias.

```python
# Estimación de las constantes u y d
u = datos.mean()[activo] + datos.std()[activo]
d = datos.mean()[activo] - datos.std()[activo]
print(f'u = {round(u,4)}\nd = {round(d,4)}')
```

Finalmente, realizamos las gráficas de las simulaciones en cada uno de los casos: 7, 10 y 30 días.

```python
# Gráfica de 20 escenarios del precio dentro de 7 días
graficar_caminata(simular_caminata(n_pasos = 7, n_sims=20, pos_inicial=S0, l_der=u, l_izq=-d, p = p))
```

```python
# Gráfica de 20 escenarios del precio dentro de 10 días
graficar_caminata(simular_caminata(n_pasos = 10, n_sims=20, pos_inicial=S0, l_der=u, l_izq=-d, p = p))
```

```python
# Gráfica de 20 escenarios del precio dentro de 30 días
graficar_caminata(simular_caminata(n_pasos = 30, n_sims=20, pos_inicial=S0, l_der=u, l_izq=-d, p = p))
```
